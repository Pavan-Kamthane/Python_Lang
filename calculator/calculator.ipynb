```python
from PyQt5.QtWidgets import *
from PyQt5 import QtCore, QtGui,QtWidgets,uic
from PyQt5.QtGui import *
from PyQt5.QtCore import *
import sys
```

```python
def action_equal(): 
    equation = calc.display.text() 
    try:     
        ans = eval(equation) 
  
        calc.display.setText(str(ans)) 
  
    except:
             
        calc.display.setText("Wrong Input")
        
def action_plus():
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display.text())
    text=doc.toPlainText()
    calc.display.setText(text + " + ") 
def action_minus():
    
    text = calc.display.text() 
    calc.display.setText(text + " - ") 
def action_div():
    
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display.text())
    text=doc.toPlainText()
    calc.display.setText(text + " / ")
def action_mul():
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display.text())
    text=doc.toPlainText()
    calc.display.setText(text + " * ")   
def action_point():
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display.text())
    text=doc.toPlainText()
    calc.display.setText(text + " . ")  
def action0(): 
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display.text())
    text=doc.toPlainText()
    calc.display.setText(text + "0")  
def action1(): 
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display.text())
    text=doc.toPlainText()
    calc.display.setText(text + "1")  
def action2():
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display.text())
    text=doc.toPlainText()
    calc.display.setText(text + "2")

def action3():  
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display.text())
    text=doc.toPlainText()
    calc.display.setText(text + "3")  
def action4():
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display.text())
    text=doc.toPlainText()
    calc.display.setText(text + "4")   
def action5():
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display.text())
    text=doc.toPlainText()
    calc.display.setText(text + "5")   
def action6(): 
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display.text())
    text=doc.toPlainText()
    calc.display.setText(text + "6")   
def action7(): 
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display.text())
    text=doc.toPlainText()
    calc.display.setText(text + "7")   
def action8(): 
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display.text())
    text=doc.toPlainText()
    calc.display.setText(text + "8")  
def action9():  
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display.text())
    text=doc.toPlainText()
    calc.display.setText(text + "9")  
def action_clear():
    calc.display.setText("")   
def action_del():
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display,text())
    text=doc.toPlainText()  
    print(text[:len(text)-1]) 
    calc.display.setText(text[:len(text)-1])
def sq():
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display.text())
    text=doc.toPlainText()
    calc.display.setText(text + "**") 
def sqrt():
    doc=QtGui.QTextDocument()
    doc.setHtml(calc.display.text())
    text=doc.toPlainText()
    calc.display.setText(text + "**0.5")
```

```python
App=QtWidgets.QApplication([])
calc=uic.loadUi('calc.ui')
calc.show()
calc.sub.clicked.connect(action_minus) 
calc.answer.clicked.connect(action_equal) 
calc.q0.clicked.connect(action0) 
calc.q1.clicked.connect(action1) 
calc.q2.clicked.connect(action2) 
calc.q3.clicked.connect(action3) 
calc.q4.clicked.connect(action4) 
calc.q5.clicked.connect(action5) 
calc.q6.clicked.connect(action6) 
calc.q7.clicked.connect(action7) 
calc.q8.clicked.connect(action8) 
calc.q9.clicked.connect(action9) 
calc.div.clicked.connect(action_div) 
calc.x.clicked.connect(action_mul) 
calc.add.clicked.connect(action_plus) 
calc.point.clicked.connect(action_point) 
calc.clr.clicked.connect(action_clear)
calc.sq.clicked.connect(sq)
calc.sqrt.clicked.connect(sqrt)

sys.exit(App.exec_())
```
